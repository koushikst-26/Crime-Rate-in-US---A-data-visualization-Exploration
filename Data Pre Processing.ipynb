```python
import pandas as pd
```

```python
crime_data = pd.read_excel('2019_Crime_Rates.xlsx')
```

```python
crime_data.head()
```

```python
crime_data.columns
```

```python
crime_columns = ['Violent Crime','Murder and Non Negligent Man Slaughter', 'Rape', 'Robbery', 'Aggrevated Assault', 
                 'Property \ncrime', 'Burglary', 'Larceny-theft', 'Motor \nvehicle \ntheft']
```

```python
for column in crime_columns:
    new_column_name = column + '_per_100k'
    crime_data[new_column_name] = ((crime_data[column] / crime_data['Population']) * 100000).round().astype(int)
```

```python
crime_data.head()
```

```python
output_file_path = '2019_Crime_rates_per100k.xlsx'
crime_data.to_excel(output_file_path, index=False)
```
